```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
File="./Data/data.csv"#brest cancer detection model
df=pd.read_csv(File)
```

```python
df.to_csv("./Data/Removed_Null_diagnosis_dataset.csv")
```

```python
df.info()
```

```python
df.drop(["Unnamed: 32","id"],axis=1,inplace=True)
```

```python
df.head()
```

```python
df["diagnosis"] = [1 if value == "M" else 0 for value in df["diagnosis"]]
df.head()
```

```python
df["diagnosis"].value_counts().plot(kind="bar")
```

```python
df.groupby(["diagnosis"])["diagnosis"].count()
```

```python
#divide the data into target variables and predicters
y=df["diagnosis"]
X=df.drop("diagnosis",axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_scaled=scaler.fit_transform(X)
```

```python
#split our data into train and test

from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X_scaled,y,test_size=0.3,random_state=42)
```

```python
#train the model
from sklearn.linear_model import LogisticRegression

model=LogisticRegression()
#train the model from data

model.fit(X_train,y_train)
#predict the target variable from test data
y_pred=model.predict(X_test)
```

```python
#evaluation of the model
from sklearn.metrics import accuracy_score

accuracy=accuracy_score(y_test,y_pred)
print(f"Accuracy: {accuracy: .2f}")
```

```python
from sklearn.metrics import classification_report

#priscision

print(classification_report(y_test,y_pred))
```

```python
import joblib

joblib.dump(model,"./model/model1.pkl")
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from Lasso_Regression import LassoRegression
```

```python
X,y=datasets.make_regression(n_samples=200,n_features=1,noise=20,random_state=42)
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
fig=plt.figure(figsize=(8,6))
plt.scatter(X[:,0],y,s=30)
plt.show()
```

```python
lassomodel=LassoRegression(lamb=1)
lassomodel.fit(X_train,y_train)
```

```python
# y_pred_line=lassomodel.pridict(X_test)
```

```python
# Prediction
y_pred_line = lassomodel.predict(X)

# Plotting
cmap = plt.get_cmap('viridis')
fig = plt.figure(figsize=(8, 6))
m1 = plt.scatter(X_train, y_train, color=cmap(0.9), s=10, label="Train")
m2 = plt.scatter(X_test, y_test, color=cmap(0.5), s=10, label="Test")
plt.plot(X, y_pred_line, color="black", label="Prediction")

plt.legend()
plt.title("Lasso Regression Fit")
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
y_pred=lassomodel.predict(X_test)
print(mean_squared_error(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(r2_score(y_test,y_pred))
```
